## Charts from meaningful data

### Bar chart

```julia
# using Plots
# using CSV
# d = CSV.read("D:notebooks/juliaplot/london2018.csv")
# bar(d.Month,d.Tmax)

using HTTP,CSV,DataFrames,Plots
data_url = "https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv"
my_file = CSV.File(HTTP.get(data_url).body)
d = DataFrame(my_file)
bar(d.Month,d.Tmax)
```

```julia
using HTTP,CSV,DataFrames,PlotlyJS
data_url = "https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv"
my_file = CSV.File(HTTP.get(data_url).body)
d2 = DataFrame(my_file)

plot(d2, x=:Month, y=:Tmax, kind="bar")
```

### Line charts

```julia
plot(d.Month, d.Tmax)
```

```julia
plot(d2, kind="scatter", mode="lines", x=:Month, y=:Tmax)
```

```julia
plot(d.Month, [d.Tmax, d.Tmin], label=["Tmax","Tmin"])
```

```julia
using HTTP,CSV,DataFrames,PlotlyJS
data_url = "https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv"
my_file = CSV.File(HTTP.get(data_url).body)
d2 = DataFrame(my_file)

# Create traces
trace1 = scatter(x=d2.Month, y=d2.Tmax,
                    mode="lines",
                    name="Tmax")
trace2 = scatter(x=d2.Month, y=d2.Tmin,
                    mode="lines",
                    name="Tmin")
plot([trace1, trace2])
```

### Scatter chart

```julia
# A scatter chart is often used to see if a pattern can be detected in the data. Here we plot the maximum temperature against the hours of sunshine. As you might expect there is a pattern: a visible correlation — the more hours of sunshine there are, the higher the temperature.
scatter(d.Tmax, d.Sun)
```

```julia
using HTTP,CSV,DataFrames,PlotlyJS
data_url = "https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv"
my_file = CSV.File(HTTP.get(data_url).body)
d2 = DataFrame(my_file)

plot(scatter(d2,x=:Tmax, y=:Sun, mode="markers"))
```

### Histogram

```julia
# d2 = CSV.read("londonweather.csv")
using HTTP,CSV,DataFrames,Plots
data_url = "https://raw.githubusercontent.com/alanjones2/dataviz/master/londonweather.csv"
my_file = CSV.File(HTTP.get(data_url).body)
d2 = DataFrame(my_file)
histogram(d2.Rain, label="Rainfall")
```
